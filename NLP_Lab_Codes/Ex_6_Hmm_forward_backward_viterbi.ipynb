```python
words=[0,1,0]
transition=[[0.7,0.3],[0.4,0.6]]
emission=[[0.1,0.4,0.5],[0.6,0.3,0.1]]
```

```python
def forward():
  initial=[0.6,0.4]
  for i in range(len(words)):
    obs=[0,0]
    for j in range(len(initial)):
          for k in range(len(initial)):
              obs[j]+=initial[k]*transition[k][j]*emission[k][words[i]]
    initial=[obs[0],obs[1]]
  print(initial[0]+initial[1])
forward()
```

```python
def backward(words):
  words=words[::-1]
  initial=[1,1]
  for i in range(len(words)):
    obs=[0,0]
    for j in range(len(initial)):
          for k in range(len(initial)):
              obs[j]+=initial[k]*transition[j][k]*emission[j][words[i]]
    initial=[obs[0],obs[1]]
  print(0.6*initial[0]+0.4*initial[1])
backward(words)
```
